```python
import matplotlib.pyplot as plt
import numpy as np

def createCluster(N,k):
    np.random.seed(10)
    points = N/k
    X = []
    for i in range(k):
        incomeCentroid = np.random.uniform(20000,200000)
        ageCentroid = np.random.uniform(20,70)
        for j in range(int(points)):
            income = np.random.normal(incomeCentroid,10000)
            age = np.random.normal(ageCentroid,2)
            X.append([income,age])
    return np.array(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale

data = createCluster(100, 5)
scaled_data = scale(data)
model = KMeans(n_clusters=3)
model = model.fit(scaled_data)

print(model)

plt.scatter(data[:,0],data[:,1],c=model.labels_)
```
