# MATPLOTLIB

```python
%matplotlib inline
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3,3,0.001)

plt.plot(x, norm.pdf(x))
plt.show()
```

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1,0.5))
plt.plot(x,norm.pdf(x,2,0.3))
plt.show()
```

```python
axes = plt.axes()
axes.set_xlim([-4,4])
axes.set_ylim([-1,2])
axes.set_xticks(np.arange(-4,4,0.5))
axes.set_yticks(np.arange(-1,2,0.5))
axes.grid()

plt.xlabel("Greebles")
plt.ylabel("Probabo")
plt.title("Mordon")

plt.plot(x,norm.pdf(x),"g-.")
plt.plot(x,norm.pdf(x,1,0.5),"b--")
plt.plot(x,norm.pdf(x,2,0.3),"r:")

plt.legend(["Turbo","Frambo","Marblo"], loc=4)
plt.show()
```

```python
plt.rcdefaults()

values = [2,63,6,8,41,3]
labels = ["Swish","Turfan","Prouder","Louder","Crakkko","Tirvian"]
explode = [0.2,0,0.4,0,0,0.5]
colors = ["r","g","b","y","m","c"]

plt.pie(values,colors=colors,labels=labels,explode=explode)
plt.title("Fantasy screeeming")
plt.show()
```

```python
plt.bar(range(0,6),values,color=colors)
plt.show()
```

# SEABORN

```python
%matplotlib inline

import pandas as pd

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")
df.head()
```

```python
gear_counts = df["# Gears"].value_counts()
gear_counts.plot(kind="bar")
```

```python
import seaborn as sns
sns.set()
gear_counts.plot(kind="bar")
```

```python
sns.distplot(tuple(df["CombMPG"]))
```

```python
df2 = df[["Cylinders","CityMPG","HwyMPG","CombMPG"]]
sns.pairplot(df2,hue="Cylinders",height=2.5)
```

```python
sns.scatterplot(x="Eng Displ",y="CombMPG",data=df)
```

```python
sns.jointplot(x="Eng Displ",y="CombMPG",data=df)
```

```python
sns.lmplot(x="Eng Displ",y="CombMPG",data=df)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.boxplot(x="Mfr Name",y="CombMPG", data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
ax = sns.swarmplot(x="Mfr Name",y="CombMPG",data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
ax = sns.countplot(x="Mfr Name",data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
df2 = df.pivot_table(index="Cylinders",columns="Eng Displ",values="CombMPG",aggfunc="mean")
sns.heatmap(df2)
```
